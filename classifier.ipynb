```python
import pandas as pd
from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype
```

```python
df = pd.read_csv('diabetes.csv')
print(df.columns)
df
```

```python
df.fillna(df.mode())
df.columns = df.columns.str.replace(' ','_')
df.replace(' ', '_', regex=True)
print(df.columns)
df
```

```python
# copy the data
print(df.columns)
df_max_scaled = df.copy()

# apply normalization techniques on Column 1
columns = ['BMI','Mental_Health','Physical_Health','Age']
for column in columns:
    if is_numeric_dtype(column):
        df_max_scaled[column] = df_max_scaled[column] /df_max_scaled[column].abs().max()
  
# view normalized data
display(df_max_scaled)
```

```python
categorical_columns = ['General_Health', 'Sex','Education','Income']
for column in categorical_columns:
    tempdf = pd.get_dummies(df[column], prefix=column)
    df = pd.merge(
        left=df,
        right=tempdf,
        left_index=True,
        right_index=True,
    )
    df = df.drop(columns=column)
df
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, test_size=0.2)
test, validation = train_test_split(test, test_size=0.5)

train_labels = train['Diabetes_binary']
train.drop('Diabetes_binary', axis=1, inplace=True)

test_labels = test['Diabetes_binary']
test.drop('Diabetes_binary', axis=1, inplace=True)

validation_labels = validation['Diabetes_binary']
validation.drop('Diabetes_binary', axis=1, inplace=True)
```

```python
dtrain = xgb.DMatrix(train, label=train_labels)
dtest = xgb.DMatrix(test, label=test_labels)
dval = xgb.DMatrix(validation, label=validation_labels)
```

```python
param = {'max_depth': 4, 'Learning_rate': 0.1, 'N_estimator': 200, 'Subsample': 0.5, 'Colsample_bytree': 1, 'Random_seed': 123, 'eval_metric': 'auc', 'Verbosity': 1}
```

```python
evallist = [(dval, 'eval'), (dtrain, 'train')]
```

```python
num_round = 20
bst = xgb.train(param, dtrain, num_round, evallist, early_stopping_rounds= 10)
bst.save_model('0001.model')
```

```python
# predict = bst.predict(dtest, iteration_range=(0, bst.best_iteration + 1))
predict = bst.predict(dtest)
```

```python
print(predict)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_recall_fscore_support
```

```python
def calculate_metrics(real_labels, predictions):
    predictions = [round(value) for value in predict]
    # print(predictions)
    accuracy = accuracy_score(real_labels, predictions)
#     print("Accuracy: %.2f%%" % (accuracy * 100.0))
    return (accuracy * 100.0), confusion_matrix(real_labels, predictions), precision_recall_fscore_support(real_labels, predictions, average='macro')
```

```python
accuracy, confusion, precision_recall_fscore = calculate_metrics(test_labels, predictions)
print("Accuracy: %.2f%%" % accuracy)
print(confusion)
print(precision_recall_fscore)
```
